```python
#STEP 48
#ANALYZE DURATION COLUMN 
df["duration"]
```

```python
#STEP 49
#SEPARATE MOVIES AND TV SHOWS
#df[type] parameter used
movies = df[df["type"] == "Movie"]
tv_shows = df[df["type"] == "TV Show"]
movies.head()
tv_shows.head()
#IT WILL SEPARATE MOVIES AND TV SHOWS
```

```python
#STEP 50
#AVERAGE MOVIE DURATION
# Check the data type and handle accordingly
if movies["duration"].dtype=="object":
    movies["duration"]=movies["duration"].str.replace("min","").astype(int)
else:
    pass

movies["duration"].mean()
#it will give u the average mean

```

```python
#STEP 51
#MOST COMMON DURATION
df['duration'].value_counts().idxmax()

```

```python
#STEP 52
#VISUALIZE MOVIE DURATION USING HISTOGRAM
#HISTOGRAM DOES NOT COMPARE CATEGORIES
#AUTOMATICALLY GROUPS NOS
plt.hist(movies["duration"], bins=20)
plt.title("Distribution of Movie Duration")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

```python
#STEP 53
#INSIGHTS
# Insight 1: Most Netflix movies have a duration between 80 and 120 minutes.

# Insight 2: Very short and very long movies are less common.

# Insight 3: The movie duration distribution is concentrated around the average duration.
```
